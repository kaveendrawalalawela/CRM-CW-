```python
close_df = data["Close"]

missing_info = []

for ticker in close_df.columns:
    col = close_df[ticker]
    total_rows = len(col)
    missing_rows = col.isna().sum()
    available_rows = total_rows - missing_rows
    missing_pct = (missing_rows / total_rows) * 100

    if missing_rows == total_rows:
        status = "Fully Missing"
    elif missing_rows > 0:
        status = "Partially Missing"
    else:
        status = "Complete"

    missing_info.append([ticker, status, missing_rows, available_rows, total_rows, missing_pct])

# Create a table
missing_table = pd.DataFrame(
    missing_info,
    columns=["Ticker", "Status", "Missing Rows", "Available Rows", "Total Rows", "Missing %"]
)

# Display the table
missing_table
```

```python
import pandas as pd

# Identify fully missing columns from missing_table
fully_missing = missing_table[missing_table['Status'] == 'Fully Missing']['Ticker'].tolist()

# Drop the fully missing columns from close_df
# Use errors='ignore' to prevent error if a column somehow doesn't exist
clean_close = close_df.drop(columns=fully_missing, errors='ignore')

# Fill any remaining NaN values using forward fill then backward fill
clean_close = clean_close.ffill().bfill()
```

```python
# Part B

# STEP 1 â€” Stock daily returns
daily_returns = clean_close.pct_change().dropna()

# Define start and end dates for market data download
start = "2022-01-01"
end = "2025-01-01"

# STEP 2 â€” Market index (S&P 500)
sp500_index = yfqa.download("^GSPC", start=start, end=end)["Close"]

# STEP 3 â€” Market daily returns
sp500_returns = sp500_index.pct_change().dropna()

# STEP 4 â€” Align dates
aligned = daily_returns.join(sp500_returns.rename(columns={'^GSPC': 'SP500'}), how="inner")

# STEP 5 â€” Calculate Beta
beta_values = {}
market_std = aligned["SP500"].std()

for ticker in daily_returns.columns:
    correlation = aligned[ticker].corr(aligned["SP500"])
    stock_std = aligned[ticker].std()
    beta = correlation * (stock_std / market_std)
    beta_values[ticker] = beta

beta_df = pd.DataFrame.from_dict(beta_values, orient="index", columns=["Beta"])

# Show all rows instead of .head()
pd.set_option("display.max_rows", None)
beta_df
```

```python
import numpy as np
from scipy.cluster.hierarchy import linkage, fcluster
from sklearn.metrics import silhouette_score

# 1. Data and linkage
X = beta_df[['Beta']].values
Z = linkage(X, method='ward')

# 2. K from dendrogram (largest jump)
distances = Z[:, 2]
gaps = np.diff(distances)
idx = np.argmax(gaps)
n = X.shape[0]
k_dendrogram = n - (idx + 1)
print(f"K from Dendrogram (largest height jump): {k_dendrogram}")

# 3. K from silhouette (global maximum in range 2â€“10)
k_values = range(2, 11)
sil_scores = {}

best_k_sil = None
best_score = -1

for k in k_values:
    labels = fcluster(Z, k, criterion='maxclust')
    score = silhouette_score(X, labels)
    sil_scores[k] = score
    print(f"K = {k}, Silhouette score = {score:.4f}")
    if score > best_score:
        best_score = score
        best_k_sil = k

print(f"\nK from Silhouette (global max): {best_k_sil} with score {best_score:.4f}")

# 4. Evaluate K = 3 as a practical / interpretable choice
k_practical = 3
score_k3 = sil_scores[k_practical]
ratio = score_k3 / best_score

print(f"\nSilhouette score at K = 3: {score_k3:.4f}")
print(f"K = 3 keeps {ratio*100:.1f}% of the best silhouette score (K = {best_k_sil}).")

# 5. Final rule:
#    - Dendrogram suggests a very coarse split (K = 2)
#    - Silhouette prefers many small clusters (K = {best_k_sil})
#    - K = 3 is:
#        * between 2 and {best_k_sil}
#        * still has a high silhouette score
#        * easy to interpret as Low / Medium / High beta
if k_practical >= k_dendrogram and k_practical <= best_k_sil and ratio >= 0.9:
    final_k = k_practical
    reason = (
        f"We choose K = {final_k} as a compromise: "
        f"it is between the dendrogram K ({k_dendrogram}) and silhouette K ({best_k_sil}), "
        f"and its silhouette score ({score_k3:.4f}) is {ratio*100:.1f}% of the maximum. "
        "K = 3 also has a clear financial interpretation (low / medium / high beta clusters)."
    )
else:
    # fallback if the condition somehow fails
    final_k = best_k_sil
    reason = f"We fallback to silhouette-optimal K = {final_k}."

print(f"\nFinal chosen K: {final_k}")
print("Explanation:", reason)

# 6. Assign clusters using final K
clusters = fcluster(Z, final_k, criterion='maxclust')
beta_df['Cluster'] = clusters
beta_df.head()
```
